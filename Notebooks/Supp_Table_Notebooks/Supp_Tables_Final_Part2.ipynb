### Supp Table 4A. Signature scores and CIBERSORTx fractions for IMM151 bulk RNA-Seq data. Clinical metadata removed from final supplementary tables to avoid republishing of data from original studies. <a name="1"></a>

```r
clin_imm151_full = readRDS("../../Trial_Cohorts/IMM151/IMM151_Clinical_Metadata_with_SigScores_CIBERSORTx.RDS")

df = data.frame(table(colnames(clin_imm151_full)))
subset(df, Freq > 1)

clin_imm151_full = subset(clin_imm151_full, select = -c(IFIT3) )
```

```r
colnames(clin_imm151_full)[1:40]
```

```r
imm151_features_clin = colnames(clin_imm151_full)[which(colnames(clin_imm151_full) == "PATIENT_ID"):which(colnames(clin_imm151_full) == "TFE3_OR_TFEB_FUSION")]
imm151_features_clin
```

### Read in signatures of interest

```r
other_sigs = readRDS(file = "../../Signatures/Final_Manuscript_Signatures_CTS_ISGs.RDS")

cts_isg_sigs = other_sigs[grep("manuscript",names(other_sigs),value = T)]
cts_isg_sigs
```

```r
x <- scan("../../Tables/Tables_Final//Supp_Table2_Signatures.txt", what="", sep="\n")
# Separate elements by one or more whitepace
y <- strsplit(x, "[[:space:]]+")
# Extract the first vector element and set it as the list element name
names(y) <- sapply(y, `[[`, 1)
#names(y) <- sapply(y, function(x) x[[1]]) # same as above
# Remove the first vector element from each list element
y <- lapply(y, `[`, -1)
y <- lapply(y, `[`, -1)

supp_table2_sigs = y
length(supp_table2_sigs)
head(supp_table2_sigs)
```

```r
names(supp_table2_sigs)
```

```r
imm151_features_sigs = intersect(c(names(supp_table2_sigs),names(cts_isg_sigs)),colnames(clin_imm151_full))
imm151_features_sigs
```

```r
imm151_features_cibersortx = grep("_CS",colnames(clin_imm151_full),value = T)
imm151_features_cibersortx = setdiff(imm151_features_cibersortx, c("P_value_CS","Correlation_CS","RMSE_CS"))
imm151_features_cibersortx
```

```r
clin_imm151_meta = clin_imm151_full[,c(imm151_features_clin,imm151_features_sigs,imm151_features_cibersortx)]
head(clin_imm151_meta)
```

```r
colnames(clin_imm151_meta)
```

### Replace informal CTS IFN signaling signature names

```r
cts_oldnames = c("manuscript_IFN1_myeloid","manuscript_IFN1_t_nk","manuscript_IFN1_tumor","manuscript_IFNG_myeloid","manuscript_IFNG_t_nk","manuscript_IFNG_tumor","manuscript_myeloid_combined","manuscript_t_nk_combined","manuscript_tumor_combined")
cts_newnames = c("Myeloid_IFNab_signaling","T_NK_IFNab_signaling","Tumor_IFNab_signaling","Myeloid_IFNg_signaling","T_NK_IFNg_signaling","Tumor_IFNg_signaling","Myeloid_IFNab_IFNg_signaling","T_NK_IFNab_IFNg_signaling","Tumor_IFNab_IFNg_signaling")
```

```r
clin_imm151_meta_names = colnames(clin_imm151_meta)

replace(clin_imm151_meta_names, clin_imm151_meta_names %in% cts_oldnames, cts_newnames)
```

```r
colnames(clin_imm151_meta) = replace(clin_imm151_meta_names, clin_imm151_meta_names %in% cts_oldnames, cts_newnames)
```

```r
head(clin_imm151_meta)
colnames(clin_imm151_meta)
```

### Read in expanded ISG signature scores

```r
imm151_exp = readRDS(file = "../../Trial_Cohorts/IMM151/IMM151_Clinical_Metadata_with_SigScores_CIBERSORTx_ExpandedISGs.RDS")
dim(imm151_exp)
corner(imm151_exp)
```

```r
expanded_sig_names = c("manuscript_IFNG_myeloid","manuscript_IFN1_myeloid","manuscript_myeloid_combined","manuscript_IFNG_tumor","manuscript_tumor_combined")
expanded_sig_names
```

```r
imm151_exp_sigs = imm151_exp[,expanded_sig_names]
dim(imm151_exp_sigs)
head(imm151_exp_sigs)
```

```r
all.equal(row.names(imm151_exp_sigs),row.names(clin_imm151_meta))
```

```r
colnames(imm151_exp_sigs) = c("Myeloid_IFNg_signaling_expanded","Myeloid_IFNab_signaling_expanded","Myeloid_IFNab_IFNg_signaling_expanded","Tumor_IFNg_signaling_expanded","Tumor_IFNab_IFNg_signaling_expanded")
head(imm151_exp_sigs)
```

```r
clin_imm151_meta = cbind(clin_imm151_meta, imm151_exp_sigs)
```

```r
write.table(clin_imm151_meta, file = "../../Tables/Tables_Final/Supp_Table4A_IMM151_Data.txt", sep = '\t',quote = F, col.names = NA)
```

### Supp Table 4B. Supp Table 4B. 6p21 copy number calls, signature scores, and CIBERSORTx fractions for JAV101 bulk RNA-Seq data. Clinical metadata removed from final supplementary tables to avoid republishing of data from original studies. <a name="2"></a>

```r
clin_jav101_full = readRDS("../../Trial_Cohorts/JAV101/JAV101_Clinical_Metadata_with_SigScores_CIBERSORTx.RDS")

df = data.frame(table(colnames(clin_jav101_full)))
subset(df, Freq > 1)

clin_jav101_full = subset(clin_jav101_full, select = -c(CD8A,Eosinophils,IFIT3, Neutrophils) )
```

```r
cn = scan("../../Trial_Cohorts/JAV101/Javelin_6p21_loss_GISTIC.txt", what = "character")
corner(cn)
length(cn)
```

```r
corner(clin_jav101_full$ID)
```

```r
cn = str_replace_all(cn, pattern = "_T","")
corner(cn)
length(cn)

length(intersect(cn, clin_jav101_full$ID))
```

```r
clin_jav101_full = mutate(clin_jav101_full, GISTIC_6p21_loss = ifelse(ID %in% cn, "Deletion","Disomy"))
table(clin_jav101_full$GISTIC_6p21_loss)
```

```r
colnames(clin_jav101_full)[1:80]
```

```r
jav101_features_clin = colnames(clin_jav101_full)[which(colnames(clin_jav101_full) == "Cohort"):which(colnames(clin_jav101_full) == "wGII")]
jav101_features_clin = c(jav101_features_clin,"GISTIC_6p21_loss","IMDC","MSKCC","Tumor_Location","Metastasis_Location","Age","Sex","PDL1_status_v1")
jav101_features_clin
```

```r
clin_jav101_meta = clin_jav101_full[,c(jav101_features_clin,jav101_features_sigs,jav101_features_cibersortx)]
head(clin_jav101_meta)
```
